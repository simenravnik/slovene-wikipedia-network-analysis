```python
summaries_list = list(summaries.values())
tfidf = TfidfVectorizer().fit_transform(summaries_list)
```

```python
idx_to_id_map = dict()
id_to_idx_map = dict()
for idx, id in enumerate(list(summaries.keys())):
    idx_to_id_map[idx] = id
    id_to_idx_map[id] = idx
    if id == str(title_to_id_map["Kidonija"]):
        i = idx
        print(i)
```

```python
avg_len = 0
size = 100
START = 0
for i in range(START, len(summaries_list)):
    # pageid_from = idx_to_id_map[i]
    # title_from = id_to_title_map[pageid_from]
    # graphid_from = pageid_to_id_map[pageid_from]

    pairwise_similarity = tfidf[i] * tfidf.T
    similarity_vector = pairwise_similarity.toarray()[0]

    idxs = np.where(similarity_vector > 0.15)[0]
    avg_len += len(idxs)
    new_edges = map(lambda idx: (i, idx), idxs)

    edges.update(new_edges)

    if i == START + size:
        break
    # print(count)
```

```python
def get_edges(i, thres=0.15):
    pairwise_similarity = tfidf[i] * tfidf.T
    similarity_vector = pairwise_similarity.toarray()[0]

    idxs = np.where(similarity_vector > thres)[0]
    return list(map(lambda idx: (i, idx), idxs))
```
